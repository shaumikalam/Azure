## Exercise 2: Running SQL Commands

Azure Databricks notebooks natively support SQL alongside Python. You can switch the language of any cell using **magic commands** — special directives that start with `%`. For example, `%sql` tells Databricks to interpret the cell as SQL.

**Scenario:** The CityMoves Transit operations team works primarily with SQL. You want to demonstrate that SQL works seamlessly in the same notebook where you write Python.

**Your task:** In the cell below, use the `%sql` magic command to write a SQL statement that:

1. Returns the current date and time.
2. Returns a label such as `'CityMoves Transit'` as a column called `system_name`.

The result should be a single-row table with at least two columns.

> 💬 **Genie Code tip:** Ask Genie Code:
> *"Write a SQL SELECT statement in a Databricks notebook that returns the current timestamp and a string literal as named columns"*
> Remember to start your cell with `%sql`.

```python
%sql
SELECT
  current_timestamp() AS current_time,
  'CityMoves Transit' AS system_name
```
